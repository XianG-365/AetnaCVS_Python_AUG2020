```python
squares = []

for i in range(1,11):
    squares.append(i**2)
squares
```

```python
[i**2 for i in range(1,11)]
```

```python
print("Squares, 1 to 10")
for square in [i**2 for i in range(1,11)]:
    print(square)
```

Warmup problem 3, using List Comprehension

```python
# original solution
results = []
for i in range(2000,3201):
    if not (i%7):
        if (i%5):
            results.append(i)
print(results)
```

```python
[i for i in range(2000,3201) if (not (i%7) and (i%5))]
```

Generator Comprehension

```python
(i for i in range(2000,3201) if (not (i%7) and (i%5)))
```

```python
for x in (i for i in range(2000,3201) if (not (i%7) and (i%5))):
    print(x,end=',')
```

```python
list((i for i in range(2000,3201) if (not (i%7) and (i%5))))
```
